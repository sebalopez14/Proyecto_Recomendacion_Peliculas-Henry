```python
movies_df['popularity'] = pd.to_numeric(movies_df['popularity'], errors='coerce')
movies_df_clean = movies_df.dropna(subset=['popularity'])
```

```python
# Eliminar valores nulos
movies_df = movies_df.dropna(subset=['release_date', 'popularity'])

# Crear un gráfico de dispersión
plt.figure(figsize=(16, 3))
sns.scatterplot(data=movies_df, x='release_date', y='popularity')
plt.title('Relación entre el Año de Lanzamiento y la Popularidad')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Popularidad')
plt.show()
```

```python
# Eliminar valores nulos
movies_df = movies_df.dropna(subset=['release_date', 'popularity'])

# Seleccionar las 10 películas más populares
top_10 = movies_df.nlargest(10, 'popularity')

# Crear un gráfico de dispersión
plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(data=top_10, x='release_date', y='popularity')

# Añadir etiquetas a cada punto
for i in range(top_10.shape[0]):
    scatter.text(
        x=top_10['release_date'].iloc[i],
        y=top_10['popularity'].iloc[i],
        s=top_10['title'].iloc[i],
        fontsize=10,
        alpha=0.7
    )

plt.title('Top 10 Películas por Popularidad')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Popularidad')
plt.xticks(rotation=45)  # Opcional, para rotar las etiquetas del eje x si están amontonadas
plt.tight_layout()  # Opcional, para ajustar el diseño y evitar que las etiquetas se corten
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.histplot(movies_df_clean['popularity'], kde=True)
plt.xscale('log')
plt.title('Distribución de la Popularidad (Escala Logarítmica)')
plt.xlabel('Popularidad (log)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
top_movies = movies_df_clean.nlargest(18, 'vote_count')

plt.figure(figsize=(20, 10))
sns.scatterplot(x='vote_count', y='popularity', data=top_movies)

for i in range(top_movies.shape[0]):
    plt.text(top_movies['vote_count'].iloc[i], top_movies['popularity'].iloc[i], 
             top_movies['title'].iloc[i], fontsize=9)

plt.title('Popularidad vs Cantidad de Votos (Top 20 Películas)')
plt.xlabel('Cantidad de Votos')
plt.ylabel('Popularidad')
plt.xscale('log')
plt.show()
```

```python
# Seleccionar solo las columnas numéricas
numerico_df = movies_df.select_dtypes(include=['float64', 'int64'])

# Calcular la matriz de correlación
matriz_correlacion = numerico_df.corr()

# Visualización de la matriz de correlación
plt.figure(figsize=(10,8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm')
plt.show()
```
